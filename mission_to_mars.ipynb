```python
from splinter import Browser
from bs4 import BeautifulSoup
import time
```

```python

executable_path = {"executable_path": "/usr/local/bin/chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
nasa_url = "https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest//"
browser.visit(nasa_url)
time.sleep(5)
html = browser.html
soup = BeautifulSoup(html, "html.parser")
#   ul > li:nth-child(1) > div > div > div.content_title > a
title = soup.find("div", class_="content_title").get_text()
paragraph = soup.find("div", class_="article_teaser_body").get_text()
browser.quit()

print(title)
print(paragraph)
```

```python
executable_path = {"executable_path": "/usr/local/bin/chromedriver"}
browser = Browser("chrome", **executable_path, headless=False)

jpl_url = "https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars"
browser.visit(jpl_url)
time.sleep(5)
html = browser.html
soup = BeautifulSoup(html, "html.parser")
browser.click_link_by_partial_text('FULL IMAGE')
time.sleep(5)
browser.click_link_by_partial_text('more info')

# <a href="/spaceimages/images/largesize/PIA17845_hires.jpg">
# <img alt="NuSTAR has provided the first observational evidence in support of a theory that says exploding stars slosh around before detonating. That theory, referred to as mild asymmetries, is shown here in a simulation by Christian Ott." title="NuSTAR has provided the first observational evidence in support of a theory that says exploding stars slosh around before detonating. That theory, referred to as mild asymmetries, is shown here in a simulation by Christian Ott." 
# class="main_image" src="/spaceimages/images/largesize/PIA17845_hires.jpg"></a>


image = soup("src")
browser.quit()

print(image)
```
